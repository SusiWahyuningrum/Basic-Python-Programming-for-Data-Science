## Dataset

```python
import pandas as pd
import numpy as np
```

## Variabel dataset




Variabel-variabel yang terdapat pada dataset ini adalah sebagai berikut:
1. **order_id** = Nomor ID pembelian(nota)
2. **order_data** = Tanggal pembelian
3. **customer_id** = Nomor ID pembeli
4. **segment** = Segmen produk 
5. **category** = Kategori produk
6. **sub_category** = Sub kategori produk
7. **sales** = Total harga pembelian
8. **quantity** = Jumlah produk dibeli
9. **profit** = Keuntungan

##1. Tentukan Segment mana dengan profit tertinggi!

```python
df1=df[df['segment']=='Consumer']
df1
```

```python
df2=df[df['segment']=='Corporate']
df2
```

```python
df3=df[df['segment']=='Home Office']
df3
```

```python
x = df1.profit.sum()
y = df2.profit.sum()
z = df3.profit.sum()
print('Consumer :',x)
print('Corporate :',y)
print('Home Office :',z)
if z<y<x :
  print('Segment Consumer adalah profit tertinggi sebesar:',x)
elif z<x<y :
  print('Segment Corporate adalah profit tertinggi sebesar:',y)
else :
  print('Segment Home Office adalah profit tertinggi sebesar:',z)
```

##2.Tentukan Category mana dengan sales terbanyak!

```python
df4=df[df['category']=='Office Supplies']
df4
```

```python
df5=df[df['category']=='Furniture']
df5
```

```python
df6=df[df['category']=='Technology']
df6
```

```python
x = df4.sales.sum()
y = df5.sales.sum()
z = df6.sales.sum()
print('Office Supplies :',x)
print('Furniture :',y)
print('Technology :',z)
if z<y<x :
  print('Category Office Supplies adalah sales terbanyak yaitu:',x)
elif z<x<y :
  print('Category Furniture adalah sales tertinggi yaitu:',y)
else :
  print('Category Technology adalah sales tertinggi yaitu:',z)
```
